```python
## Environment for bot to play around in
import gym
import gym_anytrading
from gym_anytrading.envs import TradingEnv, ForexEnv, StocksEnv, Actions, Positions
from gym_anytrading.datasets import FOREX_EURUSD_1H_ASK, STOCKS_GOOGL

#standard data libraries
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import pickle as pkl
import time

import random

import os
from collections import deque

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Input
from tensorflow.keras.layers import Flatten
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Conv1D,MaxPooling1D,GlobalAveragePooling1D
from tensorflow.keras.layers import Dropout

from tensorflow.keras.optimizers import Adam,RMSprop


class Agent():
    def __init__(self, state_size, action_size):
        #model backup
        #6
        #9 4096 4096 mem 8192 slice
        #11 5 min 
        self.weight_backup = "/tf/notebooks/11.h5"
        #how large the represented state is
        self.state_size = state_size
        #how large the set of possible actions is
        self.action_size = action_size
        #how many past datapoints the agent can "member"
        self.memory = deque(maxlen=12288)
        self.learning_rate = 1e-3
        #some weird deepq learning stuff having to do with the exploration space
        self.gamma = 0.98
        self.exploration_rate = 1
        self.exploration_min = 0.01
        self.exploration_decay = 0.995
        
        #making nn the brain
        self.brain = self._build_model()

    def _build_model(self):
        #for backups
        
        if os.path.isfile(self.weight_backup):
            model=tf.keras.models.load_model(self.weight_backup)
            model.summary()
            if os.path.isfile("./memory.pkl"):
                self.memory = pkl.load(open("./memory.pkl","rb"))
            if os.path.isfile("./exp_rate.pkl"):
                self.exploration_rate=pkl.load(open("./exp_rate.pkl","rb"))
            self.exploration_rate=self.exploration_min
            return model

        # Standard seq model
        model = Sequential()
        
        #passing in the entire state (input is of size statesize)
        model.add(Input(self.state_size))
        
        #Conv layers for "visual" data if you can say that. Conv1D decent for timeseries classification
        model.add(Conv1D(256, 32,activation='relu',padding="causal"))
        #making image /2 
        model.add(MaxPooling1D())
        #increasing filter size to 256 (more information per pixel)
        model.add(Conv1D(256, 16,activation='relu',padding="causal"))
        model.add(MaxPooling1D())
        model.add(Conv1D(256, 8,activation='relu',padding="causal"))
        #idfk used to improve generality by averaging along batches or something (helps alleviate local minima)
        #also helps flatten
        model.add(GlobalAveragePooling1D())

        # model.add(Flatten())
        
        #dense layer go brrrr
        model.add(Dense(256, activation='relu'))
        model.add(Dense(512, activation='relu'))
        model.add(Dense(512, activation='relu'))
        model.add(Dense(256, activation='relu'))
        #final action output is size of the set of actions
        model.add(Dense(self.action_size, activation='linear'))
        #rms is good. loss of mse, because i think that's a deepq learning thing, since it's trying to minimize a scalar, 
            #versus trying to minimize some onehot classification loss like in binary crossentropy
        model.compile(loss='mse',
                      optimizer=RMSprop(lr = self.learning_rate))
        
        #poops out model stuff
        model.summary()

        return model

    #for checkpoints
    def save_model(self):
        self.brain.save(self.weight_backup)
        pkl.dump(self.memory,open("./memory.pkl",'wb'))
        pkl.dump(self.exploration_rate,open("./exp_rate.pkl",'wb'))

    #random chance for random action, otherwise go with model's decision (exploration rate decreases overtime)
    def act(self, state):
        if np.random.rand() <= self.exploration_rate:
            return random.randrange(self.action_size)
        act_values = self.brain.predict(state)
        return np.argmax(act_values[0])

    #adding to that memory from before
    def remember(self, state, action, reward, next_state, done):
        self.memory.append((state, action, reward, next_state, done))

    #fitting model based on memory
    def replay(self, sample_batch_size):
        if len(self.memory) < sample_batch_size:
            return
        sample_batch = random.sample(self.memory, sample_batch_size)
        for state, action, reward, next_state, done in sample_batch:
            target = reward
            if not done:
                target = reward + self.gamma * \
                    np.amax(self.brain.predict(next_state)[0])
            target_f = self.brain.predict(state)
            target_f[0][action] = target
            self.brain.fit(state, target_f, epochs=1, verbose=0)
        if self.exploration_rate > self.exploration_min:
            self.exploration_rate *= self.exploration_decay


#setting up the data visible per iteration i think
def my_process_data(env):
    start = env.frame_bound[0] - env.window_size
    end = env.frame_bound[1]
    prices = env.df.loc[:, 'LOW'].to_numpy()[start:end]
    signal_features = env.df.loc[:, ['CLOSE', 'OPEN', 'HIGH', 'LOW','VOLUME']].to_numpy()[
        start:end]
    return prices, signal_features


class MyForexEnv(StocksEnv):
    _process_data = my_process_data
    

slice_amt=24576

class Stocks:
    def __init__(self,df=pd.read_csv("/tf/notebooks/tBTCUSD5m1546300800000-1593655395000.csv")):
        #batch size is arbitrary, but generally, larger leads to more generality i think (more things to learn from per gradient update)
        self.sample_batch_size = 64
        self.episodes = 1000
        self.df = df
        #setting up env. pretty
        #nasty spaghetti code
        self.curr_df = self.df#.loc[0:slice_amt]
        self.env = MyForexEnv(df=self.curr_df, window_size=12288,
                              frame_bound=(12288, len(self.df)),
                             )
        self.env.trade_fee_bid_percent=0.001
        self.env.trade_fee_ask_percent=0.002

        #stuff about the env and agent
        self.state_size = self.env.observation_space.shape
        self.action_size = self.env.action_space.n
        self.agent = Agent(self.state_size, self.action_size)

    #i dont think it saves until end
    def run(self):
        df_len = len(self.df.index)
        fig, ax = plt.subplots()
        skip_mod=5
        try:
            prev_score = -9999999
            for index_episode in range(self.episodes):
                fig = plt.figure()
                #setting env
                rand_start = random.randint(0,df_len-slice_amt)
                self.curr_df = self.df.loc[rand_start:rand_start+slice_amt]
                self.env = MyForexEnv(df=self.curr_df, window_size=12288,
                                      frame_bound=(12288, len(self.df)))
                state = self.env.reset()
                #idk
                state = np.reshape(state, [1, *self.state_size])
                done = False
                score = 0
                #take action get reward remember that, loop until done for that iteration(no more data points)
                
                while not done:
                    action = self.agent.act(state)
                    next_state, reward, done, info = self.env.step(action)
                    next_state = np.reshape(next_state, [1, *self.state_size])
                    score+=reward
                    self.agent.remember(
                        state, action, reward, next_state, done)
                    state = next_state
                print("Episode {}# Score: {} Exploration Rate: {}".format(index_episode, score,self.agent.exploration_rate))
                
                if score>prev_score and index_episode%skip_mod==0:
                    prev_score=score
                    self.agent.save_model()
                    
                self.env.render_all()
                plt.show()
                self.agent.replay(self.sample_batch_size)
        finally:
            #final save
            self.agent.save_model()


if __name__ == "__main__":
    stocks = Stocks()
    
    stocks.run()
    plt.cla()
    stocks.env.render_all()
    plt.show()
```
